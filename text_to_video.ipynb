```python
import torch
from diffusers import StableDiffusionPipeline
from huggingface_hub import login
import moviepy.editor as mpy
from PIL import Image
import os

# Set up Hugging Face API authentication
login("")

# Load the Stable Diffusion model
model_id = "stabilityai/stable-diffusion-2"
device = "cuda" if torch.cuda.is_available() else "cpu"

pipe = StableDiffusionPipeline.from_pretrained(model_id).to(device)

# Directory to store the generated frames
frames_dir = "generated_frames"
os.makedirs(frames_dir, exist_ok=True)

# Function to generate frames from text
def generate_frames(prompt, num_frames=10):
    images = []
    for i in range(num_frames):
        variation_prompt = f"{prompt}, frame {i+1}"
        print(f"Generating: {variation_prompt}")
        image = pipe(variation_prompt).images[0]
        image_path = f"{frames_dir}/frame_{i}.png"
        image.save(image_path)
        images.append(image_path)
    return images

# Generate frames for the animated video
text_prompt = "A futuristic city with flying cars at sunset"
num_frames = 10  # Number of frames for the video
frame_paths = generate_frames(text_prompt, num_frames)

# Function to create a video from generated frames
def create_video_from_frames(frame_paths, output_video_path, fps=2):
    frame_images = [Image.open(frame) for frame in frame_paths]
    video_clip = mpy.ImageSequenceClip([frame for frame in frame_paths], fps=fps)
    video_clip.write_videofile(output_video_path, codec='libx264')

# Create the animated video
output_video_path = "animated_video.mp4"
create_video_from_frames(frame_paths, output_video_path, fps=2)

print(f"Video saved as {output_video_path}")
```
